```python
import nltk
import newspaper
import requests
import time
from newspaper import Article

# Ensure 'punkt' is downloaded (and set NLTK data path if necessary)
nltk.download('punkt', quiet=True)
import nltk
nltk.download('punkt_tab')

# Check if 'punkt' is installed correctly
try:
    nltk.data.find('tokenizers/punkt')
    print("Punkt tokenizer is available.")
except LookupError:
    print("Punkt tokenizer not found. Please install it again.")
    exit()  # Exit if punkt is not available

# Build the newspaper object
moneycontrol = newspaper.build('http://www.cnbc.com/india', memoize_articles=False)

# Function to download an article with retries
def download_with_retry(article, retries=3, delay=5):
    for _ in range(retries):
        try:
            article.download()
            return article
        except Exception as e:
            print(f"Download failed: {e}. Retrying...")
            time.sleep(delay)
    print(f"Failed to download article after {retries} retries.")
    return None

# Open the file to write data with UTF-8 encoding, and use errors='ignore' to handle problematic characters
with open('ScrappedData_CNBC.txt', 'w', encoding='utf-8', errors='ignore') as f:
    print('\n\n\nNewspaper build finished...Awaiting Results \n\n')
    f.write('\n\n\nNewspaper build finished...Awaiting Results \n\n')

    # Print categories
    print('\n\n\nCategories are :\n\n')
    f.write('\n\n\nCategories are :\n\n')
    for category in moneycontrol.category_urls():
        print(category)
        f.write(category + '\n')

    # Print feeds
    print('\n\nFeeds are : \n')
    f.write('\n\nFeeds are : \n')
    for feed_url in moneycontrol.feed_urls():
        print(feed_url)
        f.write(feed_url + '\n')

    f.write('\n-------------------------------------------------\n\n')

    # Process each article
    index = 1
    for current_article in moneycontrol.articles:
        try:
            # Skip video pages or URLs that are known to cause issues
            if 'video' in current_article.url:
                print(f"Skipping video URL: {current_article.url}")
                continue  # Skip this article

            # Check if the URL is accessible
            response = requests.get(current_article.url)
            if response.status_code != 200:
                print(f"Article URL not accessible: {current_article.url}")
                continue  # Skip this article if URL is not accessible

            # Attempt download with retries
            current_article = download_with_retry(current_article)
            if current_article is None:
                continue  # Skip if article download failed after retries

            print(f'\n\nParsing Article {index}')
            current_article.parse()
            print(f"Title: {current_article.title}")
            f.write(f'\n\n### TITLE: \n{current_article.title}')

            print(f"\n\nText: {current_article.text}")
            f.write(f'\n\n### TEXT: \n{current_article.text}')

            try:
                # Run NLP (summary, keywords)
                current_article.nlp()
                print(f"\n\nSummary: {current_article.summary}")
                f.write(f'\n\n### Summary: \n{current_article.summary}')
                
                print(f"\n\nKeywords: {current_article.keywords}")
                f.write('\n\n### Keywords: \n')
                for keyword in current_article.keywords:
                    f.write(f'{keyword}\n')

            except Exception as e:
                print(f"Error processing NLP for article {index}: {e}")
                f.write(f"\n\nError processing NLP for article {index}: {e}\n")

            index += 1
            f.write('\n-------------------------------------------------\n\n')
            print('\n-------------------------------------------------\n\n')
        except Exception as e:
            print(f"Error processing article {index}: {e}")
            f.write(f"\n\nError processing article {index}: {e}\n")
            continue  # Skip to the next article if there's an error

```
